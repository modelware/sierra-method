# Stakeholders

```python
import requests
import pandas as pd

port = 8080
context = "fireforce6.github.io/mission-control/bundle"

query = '''
PREFIX base: <https://www.modelware.io/sierra/base#>
PREFIX stakeholder: <https://www.modelware.io/sierra/stakeholder#>
PREFIX oml: <http://opencaesar.io/oml#>
SELECT ?Stakeholder ?Category ?Description
WHERE {
    ?s a stakeholder:Stakeholder ;
      base:category ?Category ;
      base:description ?Description .
    OPTIONAL { ?s oml:label ?label }
    BIND(COALESCE(?label, STRAFTER(STR(?s), "#")) AS ?Stakeholder)
}
'''

response = requests.post(
    f"http://localhost:{port}/v0/sparql/{context}",
    data=query,
    headers={
        "Content-Type": "application/sparql-query",
        "Accept": "application/sparql-results+json"
    }
)
response.raise_for_status()
bindings = response.json()["results"]["bindings"]
rows = [{k: v["value"] for k, v in b.items()} for b in bindings]

pd.DataFrame(rows, columns=["Stakeholder", "Category", "Description"])
```
